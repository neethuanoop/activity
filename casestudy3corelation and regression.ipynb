```python
freqgraph=sales_data_add.select_dtypes(include=['float64','object'])
freqgraph.hist(figsize=(20,20))
plt.show()
```

```python

sales_data_add1=sales_data_add.drop(['Date'],axis=1)#date is dropped from the dataset
```

```python
#y-dependent
#x-independent
y=sales_data_add1['Profit(in Thousands)']
x=sales_data_add1.drop(['Profit(in Thousands)','Add4(in Thousands)'],axis=1)
```

```python
x.shape
```

```python
#splitiing data into 4 sets
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=1,test_size=0.25)
```

```python
x_train
```

```python
#creating model
from sklearn import linear_model
lr=linear_model.LinearRegression()
```

```python
model=lr.fit(x_train,y_train)#fitting data to the model
```

```python
predictions=model.predict(x_test)#predicting values for profit
```

```python
predictions
```

```python
Make_predictions=[[290,312,309,320]]#predicting value by dropping add4
```

```python
predictions=model.predict(Make_predictions)
```

```python
predictions
```

```python
from sklearn.metrics import mean_squared_error
```

```python
print('Mse is',mean_squared_error(y_test,predictions))
```
