handling outliters in ateleast 3 cloumns

```python
Q1=np.percentile(data_tit['Age'],25,interpolation='midpoint')
Q2=np.percentile(data_tit['Age'],50,interpolation='midpoint')
Q3=np.percentile(data_tit['Age'],75,interpolation='midpoint')
```

```python
print(Q1)
print(Q2)
print(Q3)
```

```python
IQR=Q3-Q1#IQr means interb quartile range
```

```python
low_lim=Q1-1.5*IQR
up_limit=Q3+1.5*IQR
```

```python
print(up_limit)
print(low_lim)
```

```python
#finding outliers
outlier=[]
for x in data_tit['Age']:
    if(x>up_limit)or(x<low_lim):
        outlier.append(x)
```

```python
outlier#all outliers present in the dataset
```

```python
data_tit.shape
```

```python
ind1=data_tit['Age']>up_limit
ageindex=data_tit.loc[ind1].index
```
