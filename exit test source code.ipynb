# handling outliers

```python
col_num=df_tra.select_dtypes(include=["int64","float64"]).columns.tolist()
fig,axs=plt.subplots(1,3,figsize=(25,8))
fig.suptitle("Box plot of numeric features to detect outliers",y=0.99,fontweight="bold",fontsize=20)
for col,ax in zip(col_num[1:],axs.flatten()):
    b=sns.boxplot(y=df_tra[col],ax=ax,color="blue")
    b.set_ylabel(col,fontsize=15)
    plt.subplots_adjust(wspace=.25)
```

```python
Q1=np.percentile(df_tra['net_yearly_income'],25,interpolation='midpoint')
Q2=np.percentile(df_tra['net_yearly_income'],50,interpolation='midpoint')
Q3=np.percentile(df_tra['net_yearly_income'],75,interpolation='midpoint')

IQR = Q3-Q1
low_lim = Q1-1.5*IQR
up_lim = Q3+1.5*IQR
```

```python
print('The values of annual income below',low_lim,'and the values  of annual income above',up_lim,'are outliers.')
```

```python
outlier=[]
for x in df_tra['net_yearly_income']:
    if x > up_lim or x < low_lim:
        outlier.append(x)
        
print('There are',len(outlier),'outliers in annual income column.','\n')
print('Outliers are:',outlier)
```

```python
len(outlier)
```

```python
ind1=df_tra['net_yearly_income'] > up_lim
i1=df_tra.loc[ind1].index

print('The index values of the outliers which are above upper limit:',i1,'\n')
```

```python
df_tra.drop(i1,inplace=True)
```

```python
plt.boxplot(df_tra['net_yearly_income'])
plt.title('Box plot of annual income')
plt.show()
```

# Label encoding

```python
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
df_tra['gender']=lb.fit_transform(df_tra['gender'])
df_tra['owns_car']=lb.fit_transform(df_tra['owns_car'])
df_tra['owns_house']=lb.fit_transform(df_tra['owns_house'])
df_tra['occupation_type']=lb.fit_transform(df_tra['occupation_type'])
```

# Modelling

```python
x = df_tra.drop(['credit_card_default','name'],axis = 1)
y = pd.DataFrame(df_tra['credit_card_default'])
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=42,test_size=0.3)
```

```python
x
```

# LOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
model = lr.fit(x_train, y_train)
y_pred = model.predict(x_test)
from sklearn.metrics import classification_report, f1_score
print(classification_report(y_pred, y_test))
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,f1_score
```

```python
print("Accuracy is:",accuracy_score(y_test,y_pred))
print("precision is",precision_score(y_test,y_pred,average='macro'))
print("recall is",recall_score(y_test,y_pred,average='macro'))
print("f1 score is",f1_score(y_test,y_pred,average='macro'))
```

```python
print(f1_score(y_pred,y_test, average='macro'))
```
