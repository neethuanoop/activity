# plot bar graph of male and female buyers participated in sales

```python
#cod for finding the count gender
#using value_count()and to_frame()
buygen=car_data['Buyer Gender'].value_counts().to_frame()
```

```python
buygen
```

```python
#code for plotting bar graph
buygen.plot.bar(color='green')
plt.title(" gender counts participated insales")
plt.xlabel('Gender')
plt.ylabel("count")
```
