```python
f,p=stats.f_oneway(sales_data_zone['Zone - A'],sales_data_zone['Zone - B'],sales_data_zone['Zone - C'],sales_data_zone['Zone - D'])
```

```python
print("p value  for significant is:",p)
```

h0:all the zones same amount of sales
h1:all the zones generate diff amount of sales

```python
if p<0.05:
    print("reject null hypothesis")
else:
    print("accept null hypothesis")
```
